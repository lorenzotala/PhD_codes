```python
import os
import glob
import pandas as pd

import numpy as np
import bokeh

import bokeh.io
import bokeh.plotting
import bokeh.palettes

from bokeh.transform import jitter
from bokeh.models import HoverTool, Range1d
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6

import seaborn as sns
import matplotlib

from scipy import stats
import math

bokeh.io.output_notebook()
```

```python
#Global variables for CSV reading
#Current directory
working_path = os.getcwd()

#Paths
data_summary_path = os.path.join(working_path, "Data_summary.csv")

#Import data
df_raw1 = pd.read_csv(data_summary_path, sep=',', na_values='*')
df_raw1['MeanRatio'] = df_raw1['MeanGreen']/df_raw1['MeanRed']
df_raw1.head()
```

```python
means = df_raw1.groupby(['Strain','MW', 'BioRep','YPosition'])['MeanRatio'].median().to_frame().reset_index()
means.head(100)
```

```python
graph_size=[1024, 512]

p = bokeh.plotting.figure(
    width=graph_size[0], 
    height=graph_size[1], 
    y_axis_label = 'Relative binding',
    title = 'Binding of Dextran 10k to WT membrane'
)

p.circle(
    source=df_raw1,
    x=jitter('YPosition', width=0.3, range=p.y_range),
    y='MeanRatio', 
    color = 'red',
    size = 5
)

p.circle(
    source=means,
    x = 'YPosition',
    y = 'MeanRatio', 
    fill_color = 'white',
    line_color = 'black',
    size = 10
)

bokeh.io.show(p)
```
