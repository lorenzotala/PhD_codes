```python
import pandas as pd
import glob, os
from tkinter import Tk     # from tkinter import Tk for Python 3.x
from tkinter.filedialog import askopenfilename
import pathlib
```

```python
def extractClassesData(root, *args):
    if len(args) == 1 and len(args) > 0:
        metadata = str(args[0])
        metadata.replace("\\","/",100)
    else:
        Tk().withdraw() # we don't want a full GUI, so keep the root window from appearing
        metadata = askopenfilename() # show an "Open" dialog box and return the path to the selected file
    folder_path = metadata.rsplit("/",1)[0]
    print(folder_path)
    Strain = folder_path.rsplit("/",1)[1]# metadata.rsplit("/",1)[1].rsplit("_",1)[1].rsplit(".",1)[0]
    print(Strain+' classes metadata:\n'+metadata)
    df_metadata = pd.read_csv(metadata)
    df_metadata['Strain'] = Strain
    class_names = df_metadata['_rlnReferenceImage'].unique()
    for n, names in enumerate(class_names):  
        class_nb = class_names[n].split('@',1)[0]
        class_nb_split = class_nb.split('0',100)
        stop_index = next((i for i, j in enumerate(class_nb_split) if j))
        class_nb_split = class_nb = class_names[n].split('@',1)[0].split('0',stop_index)
        class_int = class_nb_split[stop_index]
        df_metadata.loc[(df_metadata['_rlnReferenceImage']==class_names[n]), 'Class_nb'] = class_int
    os.chdir(folder_path)
    csv_list = []
    for file in glob.glob("*.csv"):
        csv_list.append(file)
    os.chdir(root)
    pitch = [s for s in csv_list if "pitch" in s]
    width = [s for s in csv_list if "width" in s]
    df_pitch = pd.read_csv(folder_path+"/"+pitch[0], sep = ',', na_values = '*').reset_index()
    df_pitch['Strain'] = Strain
    df_pitch = df_pitch.rename(columns={'Slice': 'Class_nb', 'Length': 'Pitch'})
    df_pitch['Pitch'] = df_pitch['Pitch']*10000
    df_width = pd.read_csv(folder_path+"/"+width[0], sep = ',', na_values = '*').reset_index()
    df_width['Strain'] = Strain
    df_width = df_width.rename(columns={'Slice': 'Class_nb', 'Length': 'Width'})
    df_width['Width'] = df_width['Width']*10000
    return df_metadata, df_pitch, df_width
```

```python
WT_path = 'C:/Users/lorenzo/Desktop/git/PhD_codes/Mechanosensation/Python_code/CryoEM/WT/metadata.csv'
t8v_path = 'C:/Users/lorenzo/Desktop/git/PhD_codes/Mechanosensation/Python_code/CryoEM/t8v/metadata.csv'
root = pathlib.Path().absolute()
print(root)
WT_metadata, WT_pitch, WT_width = extractClassesData(root, WT_path)
t8v_metadata, t8v_pitch, t8v_width = extractClassesData(root, t8v_path)

df_metadata = pd.concat([WT_metadata, t8v_metadata])
df_pitch = pd.concat([WT_pitch, t8v_pitch])
df_width = pd.concat([WT_width, t8v_width])

df_metadata.to_csv('combined_metadata.csv', index=False, encoding='utf-8-sig')
df_pitch.to_csv('combined_pitch.csv', index=False, encoding='utf-8-sig')
df_width.to_csv('combined_width.csv', index=False, encoding='utf-8-sig')
df_metadata.head()
```
