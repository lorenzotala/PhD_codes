```python
df = pd.read_csv('PolesData.csv', sep=',', na_values='*')
df['PercentTotalFluoDim']=df['TotalFluorescencePoleDim']/df['CellTotalFluorescence']
df['PercentTotalFluoBright']=df['TotalFluorescencePoleBright']/df['CellTotalFluorescence']
df['TotalPili']=df['Nb_Pili_PoleDim']+df['Nb_Pili_PoleBright']
df['TotalFlagella']=df['Nb_Flagella_PoleDim']+df['Nb_Flagella_PoleBright']
N=len(df)
y_param1='Nb_Pili_PoleDim'
y_param2='Nb_Pili_PoleBright'
N_2dim=len(df.loc[(df[y_param1]==2)]
N_1dim=len(df.loc[df[y_param1]==1])
N_0dim=len(df.loc[df[y_param1]==0])
N_2brigth=len(df.loc[df[y_param2]==2])
N_1brigth=len(df.loc[df[y_param2]==1])
N_0brigth=len(df.loc[df[y_param2]==0])
print('N='+str(N)+', N_2dim='+str(N_2dim)+', N_1dim='+str(N_1dim)+', N_0dim='+str(N_0dim))
print('N='+str(N)+', N_2brigth='+str(N_2brigth)+', N_1brigth='+str(N_1brigth)+', N_0brigth='+str(N_0brigth))
```

```python
P_2dim=N_2dim/N
P_1dim=N_1dim/N
P_0dim=N_0dim/N
P_2bright=N_2brigth/N
P_1bright=N_1brigth/N
P_0bright=N_0brigth/N

print('P(2pili|dim)='+str(P_2dim)+', P(1pili|dim)='+str(P_1dim)+', P(0pili|dim)='+str(P_0dim))
print('P(2pili|bright)='+str(P_2bright)+', P(1pili|bright)='+str(P_1bright)+', P(0pili|bright)='+str(P_0bright))
```

```python
#It doesn't mean anything!! As I already separated bright from dim poles in the analysis.
P_pole=1/2

PC_2bright=P_2bright/P_pole
PC_1bright=P_1bright/P_pole
PC_0bright=P_0bright/P_pole

PC_2dim=P_2dim/P_pole
PC_1dim=P_1dim/P_pole
PC_0dim=P_0dim/P_pole

print('P(2pili|dim)='+str(PC_2dim)+', P(1pili|dim)='+str(PC_1dim)+', P(0pili|dim)='+str(PC_0dim))
print('P(2pili|bright)='+str(PC_2bright)+', P(1pili|bright)='+str(PC_1bright)+', P(0pili|bright)='+str(PC_0bright))
```

```python
p0 = bokeh.plotting.figure(
    width=600, 
    height=600, 
    x_axis_label='# pili', 
    y_axis_type='linear',
    y_axis_label ='P(# pili|pole state)',
    title="Probability of having # pili if bright or dim pole"
)

x_1=[0,1,2]
y_1=[PC_0dim, PC_1dim, PC_2dim]

p0.line(
    x=x_1,
    y=y_1, 
    line_color = 'blue',
    #fill_color = 'blue',
    alpha=0.7,
    #legend = labelsAll[i]
    legend = 'Dim Pole'
)

y_2=[PC_0bright, PC_1bright, PC_2bright]
p0.line(
    x=x_1,
    y=y_2, 
    line_color = 'red',
    #fill_color = 'red',
    alpha=0.7,
    #legend = labelsAll[i]
    legend = 'Bright pole'
)


bokeh.io.show(p0)
```

```python
a=df.loc[(df[y_param2]==1)]
b=a['TotalFluorescencePoleDim']
c=a['TotalFluorescencePoleBright']
p = bokeh.plotting.figure(
        width=600, 
        height=600, 
        x_axis_label='totalFlu', 
        y_axis_type='linear',
        y_axis_label ='counts',
        title="Distribution of cells with 1 pilus"
    )
def make_hist_plot(b,parameter,coef,condition,color):
    hist, edges = np.histogram(list(b), density=True, bins=int(len(b)/coef))
    
    p.quad(
        top=hist, 
        bottom=0, 
        left=edges[:-1], 
        right=edges[1:],
        fill_color=color,
        line_color="white",
        alpha=0.5,
        legend=condition
    )
    return p

p_hist=make_hist_plot(b,'TotalFluorescencePoleDim',1,'Dim','navy')

bokeh.io.show(p_hist)

```
