```python
df = pd.read_csv('PolesData.csv', sep=',', na_values='*')
df['PercentTotalFluoDim']=df['TotalFluorescencePoleDim']/df['CellTotalFluorescence']
df['PercentTotalFluoBright']=df['TotalFluorescencePoleBright']/df['CellTotalFluorescence']
df['TotalPili']=df['Nb_Pili_PoleDim']+df['Nb_Pili_PoleBright']
df['TotalFlagella']=df['Nb_Flagella_PoleDim']+df['Nb_Flagella_PoleBright']

y_param1='Nb_Pili_PoleDim'
y_param2='Nb_Pili_PoleBright'
N_3dim=len(df.loc[(df[y_param1]==3)])
N_2dim=len(df.loc[(df[y_param1]==2)])
N_1dim=len(df.loc[df[y_param1]==1])
N_0dim=len(df.loc[df[y_param1]==0])
N_3brigth=len(df.loc[df[y_param2]==3])
N_2brigth=len(df.loc[df[y_param2]==2])
N_1brigth=len(df.loc[df[y_param2]==1])
N_0brigth=len(df.loc[df[y_param2]==0])
N=len(df)
print('N='+str(N)+', N_3dim='+str(N_3dim)+', N_2dim='+str(N_2dim)+', N_1dim='+str(N_1dim)+', N_0dim='+str(N_0dim))
print('N='+str(N)+', N_3brigth='+str(N_3brigth)+', N_2brigth='+str(N_2brigth)+', N_1brigth='+str(N_1brigth)+', N_0brigth='+str(N_0brigth))
```

```python
P_3dim=N_3dim/(N_3dim+N_3brigth)
P_2dim=N_2dim/(N_2dim+N_2brigth)
P_1dim=N_1dim/(N_1dim+N_1brigth)
P_0dim=N_0dim/(N_0dim+N_0brigth)
P_3bright=N_3brigth/(N_3dim+N_3brigth)
P_2bright=N_2brigth/(N_2dim+N_2brigth)
P_1bright=N_1brigth/(N_1dim+N_1brigth)
P_0bright=N_0brigth/(N_0dim+N_0brigth)

print('P(3pili|dim)='+str(P_3dim)+' P(2pili|dim)='+str(P_2dim)+', P(1pili|dim)='+str(P_1dim)+', P(0pili|dim)='+str(P_0dim))
print('P(3pili|bright)='+str(P_3bright)+' P(2pili|bright)='+str(P_2bright)+', P(1pili|bright)='+str(P_1bright)+', P(0pili|bright)='+str(P_0bright))
print('tot='+str(P_3bright+P_3dim)+' tot='+str(P_2bright+P_2dim)+', tot='+str(P_1bright+P_1dim)+', tot='+str(P_0bright+P_0dim))
```

```python
p0 = bokeh.plotting.figure(
    width=600, 
    height=600, 
    x_axis_label='# pili', 
    y_axis_type='linear',
    y_axis_label ='P(# pili|pole state)',
    title="Probability of having # pili if bright or dim pole"
)

x_1=[0,1,2,3]
y_1=[P_0dim, P_1dim, P_2dim, P_3dim]

p0.line(
    x=x_1,
    y=y_1, 
    line_color = 'blue',
    #fill_color = 'blue',
    alpha=0.7,
    #legend = labelsAll[i]
    legend = 'Dim Pole'
)

y_2=[P_0bright, P_1bright, P_2bright, P_3bright]
p0.line(
    x=x_1,
    y=y_2, 
    line_color = 'red',
    #fill_color = 'red',
    alpha=0.7,
    #legend = labelsAll[i]
    legend = 'Bright pole'
)


bokeh.io.show(p0)
```

```python
a=df.loc[(df[y_param1]==2)]
d=df.loc[(df[y_param2]==2)]
b=a['TotalFluorescencePoleDim']
c=d['TotalFluorescencePoleBright']
def make_hist_plot(b, c,coef,condition,color):
    hist, edges = np.histogram(list(b), density=True, bins=int(len(b)/coef))
    hist1, edges1 = np.histogram(list(c), density=True, bins=int(len(b)/coef))
    p = bokeh.plotting.figure(
        width=600, 
        height=600, 
        x_axis_label='totalFlu', 
        y_axis_type='linear',
        y_axis_label ='counts',
        title="Distribution of cells with 2 pilus"
    )
    p.quad(
        top=hist, 
        bottom=0, 
        left=edges[:-1], 
        right=edges[1:],
        fill_color=color[0],
        line_color="white",
        alpha=0.5,
        legend=condition[0]
    )
    p.quad(
        top=hist1, 
        bottom=0, 
        left=edges1[:-1], 
        right=edges1[1:],
        fill_color=color[1],
        line_color="white",
        alpha=0.5,
        legend=condition[1]
    )
    return p
color=['navy', 'red']
condition=['Dim', 'Bright']
p_hist=make_hist_plot(b,c,3,condition,color)


bokeh.io.show(p_hist)

```
